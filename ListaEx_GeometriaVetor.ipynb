```python
#5-A Órbita do Cometa (Tópico: Elipse)

import numpy as np
import matplotlib.pyplot as plt

# 1. Identificar os semieixos a e b

# A equação geral da elipse centrada na origem é (x^2 / a^2) + (y^2 / b^2) = 1
# Neste caso, a^2 = 100 e b^2 = 36 (assumindo que o eixo maior está no eixo x)

a_quadrado = 100
b_quadrado = 36

# Semieixos
a = np.sqrt(a_quadrado) # Semieixo maior
b = np.sqrt(b_quadrado) # Semieixo menor

print(f"1. Semieixos a e b:")
print(f"   Semieixo maior (a): {a:.2f} UA")
print(f"   Semieixo menor (b): {b:.2f} UA")

# ---

# 2. Calcular a distância focal c (onde está o sol)

# Para a elipse, vale a relação: a^2 = b^2 + c^2
# Onde 'c' é a distância do centro ao foco.

c_quadrado = a_quadrado - b_quadrado
c = np.sqrt(c_quadrado)

print(f"\n2. Distância focal c:")
print(f"   Distância focal (c): {c:.2f} UA")

# O Sol está em um dos focos. Os focos estão em (c, 0) e (-c, 0).
foco1 = (c, 0)
foco2 = (-c, 0)
sol = foco2 # Vamos assumir que o Sol está no foco F2 (-c, 0)

# ---

# 3. Calcular o periélio e o afélio

# O periélio (ponto mais próximo do Sol) e o afélio (ponto mais distante)
# são calculados em relação ao centro da elipse.

# Se o Sol (foco) está em (-c, 0) e a elipse se estende de -a a a no eixo x:
# Periélio (distância mínima): Distância do ponto (a, 0) ao foco (-c, 0) => a - c
# Afélio (distância máxima): Distância do ponto (-a, 0) ao foco (-c, 0) => a + c


# Se o Sol (foco) está em (-c, 0):
# Ponto mais próximo (Periélio) é (a, 0) -> Distância = a - c
# Ponto mais distante (Afélio) é (-a, 0) -> Distância = a + c

perielio = a - c
afelio = a + c

print(f"\n3. Periélio e Afélio:")
print(f"   Periélio (distância mínima): {perielio:.2f} UA")
print(f"   Afélio (distância máxima): {afelio:.2f} UA")

# ---

# 4. Plotar a elipse e marcar os focos

# Gerar pontos para a elipse
theta = np.linspace(0, 2 * np.pi, 360) # Ângulos de 0 a 2*pi
x_elipse = a * np.cos(theta)
y_elipse = b * np.sin(theta)

plt.figure(figsize=(8, 6))

# Plotar a elipse
plt.plot(x_elipse, y_elipse, label=r'Órbita do Cometa: $\frac{x^2}{100} + \frac{y^2}{36} = 1$', color='blue')

# Marcar o centro
plt.plot(0, 0, 'ko', label='Centro da Elipse', markersize=4)

# Marcar os focos (onde um deles é o Sol)
# Vamos marcar o Sol no foco 2 (-c, 0)
plt.plot(foco1[0], foco1[1], 'ro', label=f'Foco 1 ({foco1[0]:.2f}, 0)', markersize=7)
plt.plot(foco2[0], foco2[1], 'y*', label=f'Sol / Foco 2 ({foco2[0]:.2f}, 0)', markersize=12, markeredgecolor='black')
# Fazer um marcador extra para o Sol para ficar mais evidente
plt.plot(foco2[0], foco2[1], 'yo', markersize=10, alpha=0.5)

# Marcar Periélio e Afélio (os vértices)
plt.plot(a, 0, 'go', label=f'Periélio (x={a:.2f})', markersize=6)
plt.plot(-a, 0, 'mo', label=f'Afélio (x={-a:.2f})', markersize=6)


# Configurações do plot
plt.title('Órbita Elíptica do Cometa', fontsize=16)
plt.xlabel('Eixo x (UA)')
plt.ylabel('Eixo y (UA)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.axis('equal') # Garante que a escala dos eixos seja a mesma para a elipse não parecer distorcida
plt.legend(loc='upper right')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.show()

```

```python
#6-O "Raio Laser" (Tópico: Interseção Reta-Círculo)

import numpy as np
import matplotlib.pyplot as plt

def resolver_intersecao(h, k, R, A, B, C_linear):
    """
    Determina a intersecção entre a reta Ax + By + C_linear = 0 e o círculo
    (x-h)^2 + (y-k)^2 = R^2.

    Para o caso do problema, a reta é y = x, que pode ser reescrita como:
    1*x + (-1)*y + 0 = 0.
    Portanto, A=1, B=-1, C_linear=0.



    Parâmetros:
    h, k (float): Coordenadas do centro do círculo.
    R (float): Raio do círculo.
    # Os parâmetros A, B, C_linear são passados para manter a assinatura genérica,
    # mas o cálculo usa a equação específica do problema.
    """

    # 1. Obter os coeficientes da equação quadrática 2x^2 - 40x + 191 = 0
    # Equação geral da elipse: (x-h)^2 + (y-k)^2 = R^2
    # Reta: y = x. Substituindo:
    # 2(x - h)^2 = R^2  (Pois h=k no problema)
    # 2(x^2 - 2hx + h^2) - R^2 = 0
    # 2x^2 - 4hx + (2h^2 - R^2) = 0

    # Coeficientes da equação do problema: 2x^2 - 40x + 191 = 0
    A_quad = 2.0
    B_quad = -4.0 * h
    C_quad = 2.0 * h**2 - R**2

    # Aplicando os valores do problema: h=10, R=3
    # A_quad = 2.0
    # B_quad = -4 * 10 = -40.0
    # C_quad = 2 * 10**2 - 3**2 = 200 - 9 = 191.0

    # 2. Calcular o discriminante (Delta)
    discriminante = B_quad**2 - 4 * A_quad * C_quad

    # 3. Analisar o resultado
    print(f"\n--- Análise da Interseção ---")
    print(f"Equação Quadrática: {A_quad:.1f}x^2 + {B_quad:.1f}x + {C_quad:.1f} = 0")
    print(f"Discriminante (Δ = B² - 4AC): {discriminante:.2f}")

    if discriminante > 0:
        # Calcular os pontos de intersecção (raízes)
        x1 = (-B_quad + np.sqrt(discriminante)) / (2 * A_quad)
        x2 = (-B_quad - np.sqrt(discriminante)) / (2 * A_quad)
        # Como y = x
        ponto1 = (x1, x1)
        ponto2 = (x2, x2)

        resultado = (
            f"A reta cruza o círculo em **2 pontos** (Δ > 0).",
            2,
            ponto1,
            ponto2
        )
    elif discriminante == 0:
        x_tangente = -B_quad / (2 * A_quad)
        ponto_tangente = (x_tangente, x_tangente)
        resultado = (
            f"A reta **tangencia** o círculo em **1 ponto** (Δ = 0).",
            1,
            ponto_tangente,
            None
        )
    else: # discriminante < 0
        resultado = (
            f"A reta **NÃO toca** o círculo (Δ < 0).",
            0,
            None,
            None
        )

    return resultado

# --- Contexto do Problema ---
h, k = 10, 10 # Centro do Círculo
R = 3         # Raio do Círculo
# A, B, C_linear para a reta y = x (ou 1x - 1y + 0 = 0)
A, B, C_linear = 1, -1, 0

# Execução da função
mensagem, num_pontos, p1, p2 = resolver_intersecao(h, k, R, A, B, C_linear)

# Exibição do resultado
print("\n--- Resultado Final do Raio Laser ---")
print(f"Centro do Escudo: ({h}, {k}), Raio: {R}")
print(f"Raio Laser (Reta): y = x")
print(f"Conclusão: {mensagem}")

# ---
## Visualização (Plotagem)
# A plotagem é útil para verificar visualmente a conclusão

fig, ax = plt.subplots(figsize=(8, 8))

# Círculo (Escudo)
circulo = plt.Circle((h, k), R, color='blue', fill=False, linewidth=2, label='Escudo Circular')
ax.add_patch(circulo)

# Reta (Raio Laser)
x_reta = np.linspace(0, 15, 100)
y_reta = x_reta
ax.plot(x_reta, y_reta, 'r--', label='Raio Laser (y=x)')

# Centro do Círculo
ax.plot(h, k, 'bo', label='Centro C(10, 10)')

# Origem (Disparo)
ax.plot(0, 0, 'g*', markersize=10, label='Origem (0, 0)')

# Configurações do Plot
ax.set_xlim(0, 15)
ax.set_ylim(0, 15)
ax.set_xlabel('Eixo X')
ax.set_ylabel('Eixo Y')
ax.set_title('Intersecção do Raio Laser com o Escudo Protetor')
ax.set_aspect('equal', adjustable='box')
ax.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

```python
#9- Localização Geográfica
questão 9:
-----------------------------------------------------
# Começo do código
import numpy as np
import matplotlib.pyplot as plt
import math

# Dados do exercício
a = 3.0   # semi-eixo real
b = 4.0   # semi-eixo conjugado

# 1) Cálculo de c (distância do centro até cada foco)
# Fórmula: c^2 = a^2 + b^2  (para hipérbole)
c = math.sqrt(a**2 + b**2)

# Coordenadas dos focos e vértices
focos = [(-c, 0.0), (c, 0.0)]
vertices = [(-a, 0.0), (a, 0.0)]

# 2) Equações das assíntotas: y = ± (b/a) x
slope = b / a  # coeficiente angular das assíntotas

# Preparar dados para plotar os ramos da hipérbole
# Para uma hipérbole horizontal: x^2/a^2 - y^2/b^2 = 1  =>  y = ± (b/a) * sqrt(x^2 - a^2)
x_right = np.linspace(a + 0.01, 12, 600)   # ramo direito (x >= a)
x_left  = np.linspace(-12, -a - 0.01, 600)  # ramo esquerdo (x <= -a)

y_right_top =  (b / a) * np.sqrt(x_right**2 - a**2)
y_right_bot = -(b / a) * np.sqrt(x_right**2 - a**2)
y_left_top  =  (b / a) * np.sqrt(x_left**2 - a**2)
y_left_bot  = -(b / a) * np.sqrt(x_left**2 - a**2)

# Assíntotas (linhas retas)
x_asym = np.linspace(-12, 12, 200)
y_asym_pos = slope * x_asym
y_asym_neg = -slope * x_asym

# --- PLOT ---
plt.figure(figsize=(8,6))
# ramos da hipérbole
plt.plot(x_right, y_right_top, linewidth=1.5, label='Ramo direito (superior)')
plt.plot(x_right, y_right_bot, linewidth=1.5, label='Ramo direito (inferior)')
plt.plot(x_left,  y_left_top,  linewidth=1.5, label='Ramo esquerdo (superior)')
plt.plot(x_left,  y_left_bot,  linewidth=1.5, label='Ramo esquerdo (inferior)')

# assíntotas (tracejadas)
plt.plot(x_asym, y_asym_pos, linestyle='--', linewidth=1, label=f'Assíntota y = +{slope:.3f}x')
plt.plot(x_asym, y_asym_neg, linestyle='--', linewidth=1, label=f'Assíntota y = -{slope:.3f}x')

# marcar focos e vértices
fxs, fys = zip(*focos)
vxs, vys = zip(*vertices)
plt.scatter(fxs, fys, marker='o')   # focos
plt.scatter(vxs, vys, marker='x')   # vértices

# Anotações
for i, (xf, yf) in enumerate(focos):
    plt.annotate(f'F{i+1}({xf:.1f}, {yf:.1f})', (xf, yf), textcoords="offset points", xytext=(5,5))
for i, (xv, yv) in enumerate(vertices):
    plt.annotate(f'V{i+1}({xv:.1f}, {yv:.1f})', (xv, yv), textcoords="offset points", xytext=(5,-12))

plt.axhline(0, linewidth=0.6)
plt.axvline(0, linewidth=0.6)
plt.grid(True, linestyle=':', alpha=0.7)
plt.title(r'Hipérbole: $\dfrac{x^2}{a^2} - \dfrac{y^2}{b^2} = 1$  (a=3, b=4)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.xlim(-12, 12)
plt.ylim(-10, 10)
plt.legend(loc='upper right')

# Salvar figura
plt.savefig('hiperbole_ex9.png', dpi=150, bbox_inches='tight')
plt.show()

# Resultados numéricos
print("a =", a, "b =", b)
print("c = sqrt(a^2 + b^2) =", c)
print("Focos: ", focos)
print("Vértices: ", vertices)
print("Assíntotas: y = ±(b/a) x  => y = ±", slope, "x")
```

```python
#1- O Vetor do GPS (Tópico: Vetores)

import math

def calcular_vetor_deslocamento(ponto_a, ponto_b):
    vetor_ab = (
        ponto_b[0] - ponto_a[0],
        ponto_b[1] - ponto_a[1],
        ponto_b[2] - ponto_a[2]
    )
    magnitude = math.sqrt(
        vetor_ab[0]**2 + vetor_ab[1]**2 + vetor_ab[2]**2
    )
    return vetor_ab, magnitude

Ponto_A = (2, 3, 0)
Ponto_B = (7, 10, 0)

vetor, distancia = calcular_vetor_deslocamento(Ponto_A, Ponto_B)

print("--- 1. O Vetor do GPS ---")
print(f"Vetor de Deslocamento AB: {vetor}")
print(f"Magnitude (Distância) do Vetor: {distancia:.2f} unidades")
print("---------------------------\n")
```

```python
#7-Segurança em Drones e Estruturas
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


P1 = np.array([3.0, 0.0, 0.0])
P2 = np.array([0.0, 4.0, 0.0])
P3 = np.array([0.0, 0.0, 5.0])
# Ponto Q
Q = np.array([5.0, 5.0, 5.0])


# Funções de Álgebra Linear


def calcular_coeficientes_plano(p1, p2, p3):

    # Vetores que definem o plano
    vetor_u = p2 - p1
    vetor_v = p3 - p1

    #Calcular o produto vetorial
    vetor_normal = np.cross(vetor_u, vetor_v)
    A, B, C = vetor_normal

    # Cálculo de D: A*x1 + B*y1 + C*z1 + D = 0  => D = - (A*x1 + B*y1 + C*z1
    D = -np.dot(vetor_normal, p1)

    return A, B, C, D


def calcular_distancia_ponto_plano(ponto_q, A, B, C, D):
    x0, y0, z0 = ponto_q
    numerador = np.abs(A * x0 + B * y0 + C * z0 + D)

    # Denominador: Norma do vetor normal (||vetor_normal||)
    denominador = np.sqrt(A**2 + B**2 + C**2)



    distancia = numerador / denominador
    return distancia


def plotar_3d(p1, p2, p3, q, A, B, C, D, distancia):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111, projection='3d')

    # Plotando o Sensor
    ax.scatter(q[0], q[1], q[2], color='red', marker='o', s=150, label=f'Sensor Q({q[0]}, {q[1]}, {q[2]})')

    # Plotando a Superfície Triangular

    # Plotar os vértices
    ax.scatter(p1[0], p1[1], p1[2], color='blue', marker='^', s=80, label='Vértices P1, P2, P3')
    ax.scatter(p2[0], p2[1], p2[2], color='blue', marker='^', s=80)
    ax.scatter(p3[0], p3[1], p3[2], color='blue', marker='^', s=80)

    # Plotar as arestas do triângulo
    x_tri = [p1[0], p2[0], p3[0], p1[0]]
    y_tri = [p1[1], p2[1], p3[1], p1[1]]
    z_tri = [p1[2], p2[2], p3[2], p1[2]]
    ax.plot(x_tri, y_tri, z_tri, color='blue', linestyle='--', linewidth=1)

    # 3Preencher a superfície
    # plot_trisurf pode receber as coordenadas X, Y e Z dos vértices diretamente
    ax.plot_trisurf([p1[0], p2[0], p3[0]], [p1[1], p2[1], p3[1]], [p1[2], p2[2], p3[2]],
                    color='cyan', alpha=0.5, label='Superfície de Pouso')

    # Plotando a Distância
    # t é o fator de escala para o vetor normal: t = (A*x0 + B*y0 + C*z0 + D) / (A^2 + B^2 + C^2)
    denominador_t = A**2 + B**2 + C**2
    t = (A*q[0] + B*q[1] + C*q[2] + D) / denominador_t
    vetor_normal = np.array([A, B, C])
    Q_proj = q - t * vetor_normal

    # Plotar a linha que representa a distância mínima
    ax.plot([q[0], Q_proj[0]], [q[1], Q_proj[1]], [q[2], Q_proj[2]],
            color='green', linestyle='-', linewidth=3, label=f'Distância Mínima ({distancia:.4f})')

    # Configurações do Gráfico
    ax.set_xlabel('Eixo X (m)')
    ax.set_ylabel('Eixo Y (m)')
    ax.set_zlabel('Eixo Z (m)')
    ax.set_title('Análise de Segurança: Sensor de Drone vs. Superfície de Pouso')

    # Ajuste de limites
    todos_pontos = np.vstack([p1, p2, p3, q, Q_proj])
    max_val = np.max(todos_pontos) + 1
    min_val = np.min(todos_pontos) - 1

    ax.set_xlim([min_val, max_val])
    ax.set_ylim([min_val, max_val])
    ax.set_zlim([min_val, max_val])

    ax.legend()

    # Salvar o gráfico para o relatório
    caminho_grafico = "distancia_sensor_plano_v2.png"
    plt.savefig(caminho_grafico)

    return caminho_grafico
# main
if __name__ == "__main__":
    print("--- Iniciando Análise de Segurança em Drones (v2.0) ---")
    print(f"P1={P1}, P2={P2}, P3={P3} | Q={Q}")
    print("-" * 50)

    # Determinar a Equação do Plano
    A, B, C, D = calcular_coeficientes_plano(P1, P2, P3)
    print("1. Coeficientes do Plano (Ax + By + Cz + D = 0):")
    print(f"   Vetor Normal (A, B, C) = ({A}, {B}, {C})")
    print(f"   D = {D}")
    print(f"   Equação: {A:.0f}x + {B:.0f}y + {C:.0f}z + {D:.0f} = 0")

    # Calcular a Distância
    distancia = calcular_distancia_ponto_plano(Q, A, B, C, D)

    print("-" * 50)
    print("2. Cálculo da Distância Mínima:")
    print(f"   Distância (d) = {distancia:.4f} unidades")

    # Visualização 3D
    caminho_grafico = plotar_3d(P1, P2, P3, Q, A, B, C, D, distancia)

    print("-" * 50)
    print(f"3. Visualização 3D gerada e salva em: {caminho_grafico}")
```

```python
#8- Design de um Farol
import numpy as np
import matplotlib.pyplot as plt


A = 0.25 # Coeficiente 'a' da parábola y = a(x-h)^2 + k
H = 0.0  # Coordenada x do Vértice
K = 0.0  # Coordenada y do Vértice


#  Funções de Cálculo
def calcular_elementos_parabola(a, h, k):
    # p é a distância do Vértice ao Foco e do Vértice à Diretriz: p = 1 / (4a)
    p = 1.0 / (4.0 * a)

    # Vértice
    vertice = (h, k)

    # Foco
    foco = (h, k + p)

    # Diretriz
    diretriz_y = k - p

    return p, vertice, foco, diretriz_y


# Função de Plotagem
def plotar_parabola(a, h, k, foco, diretriz_y):


    # Geração dos pontos da parábola
    x = np.linspace(h - 5, h + 5, 400)
    y = a * (x - h)**2 + k

    # Configurar a figura e os eixos
    fig, ax = plt.subplots(figsize=(10, 8))

    # Plotar a Parábola
    ax.plot(x, y, label=f'y = {a}*(x - {h})^2 + {k}', color='blue', linewidth=2)

    # Marcar o Vértice
    ax.plot(h, k, 'o', color='green', markersize=8, label=f'Vértice V({h}, {k})')

    # Marcar o Foco
    f_x, f_y = foco
    ax.plot(f_x, f_y, 'X', color='red', markersize=10, label=f'Foco F({f_x:.2f}, {f_y:.2f})')

    # Desenhar a Diretriz
    ax.axhline(diretriz_y, color='gray', linestyle='--', label=f'Diretriz y = {diretriz_y:.2f}')

    # Configurações Finais
    ax.set_title('Parábola para Design de Farol')
    ax.set_xlabel('Eixo X')
    ax.set_ylabel('Eixo Y')
    ax.axvline(0, color='black', linewidth=0.5) # Eixo Y
    ax.axhline(0, color='black', linewidth=0.5) # Eixo X
    ax.grid(True, linestyle=':', alpha=0.6)
    ax.legend()

    # Salvar o gráfico
    caminho_grafico = "farol_parabola_simples.png"
    plt.savefig(caminho_grafico)

    return caminho_grafico


# main
if __name__ == "__main__":

    print("Análise de Parábola para Design de Farol")
    print(f"Equação de entrada: y = {A}x^2")
    print("-" * 50)

    # Calcular os elementos
    p, V, F, D_y = calcular_elementos_parabola(A, H, K)

    #  Apresentar os resultados
    print("Elementos Calculados:")
    print(f"  Parâmetro p = {p:.2f}")
    print(f"  Vértice V(h, k) = {V}")
    print(f"  Foco F(h, k + p) = {F} <-- Local da Lâmpada")
    print(f"  Diretriz: y = {D_y:.2f}")

    # Gerar Visualização
    caminho_grafico = plotar_parabola(A, H, K, F, D_y)

    print("-" * 50)
    print(f" Visualização gerada e salva em: {caminho_grafico}")
```

```python
#3-O Risco de Colisão (Tópico: Distância entre Ponto e Reta)
import math

def calcular_distancia_ponto_reta(equacao_reta, ponto):

    A, B, C = equacao_reta
    x0, y0 = ponto
    distancia = abs(A*x0 + B*y0 + C) / math.sqrt(A**2 + B**2)
    return distancia

# Exemplo:
reta = (3, 4, -12)
montanha = (10, 10)
print(calcular_distancia_ponto_reta(reta, montanha))
```

```python
#4-A Área de Cobertura do Wi-Fi (Tópico: Circunferência)
import math

def verificar_sinal(centro_router, raio, pos_usuario):
    cx, cy = centro_router
    px, py = pos_usuario
    distanciaa = math.sqrt((px - cx)**2 + (py - cy)**2)

    if distanciaa < raio:
        status = "Dentro da cobertura"
    elif distanciaa == raio:
        status = "Na borda da cobertura"
    else:
        status = "Fora da cobertura"

    return status

#Exemplo:
print(verificar_sinal((5, 5), 10, (12, 10)))
```
